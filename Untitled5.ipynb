```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/colab/UrunListesi.csv", encoding="cp1254")
```

```python
df
```

```python
df.drop("siparis_id",axis=1,inplace = True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["siparis_gecmisi"].unique()
```

```python
imp =IterativeImputer(random_state=0)
```

```python

from sklearn.impute import SimpleImputer
  
```

```python
df.replace("?", np.nan, inplace=True)  # "?" karakterini NaN ile değiştirin

imp = SimpleImputer(strategy="mean")
df_handled_missing_value = imp.fit_transform(df.loc[:, "kullanici_id":"departman_id"])
df_without_categorical_value = pd.DataFrame(df_handled_missing_value)
```

```python
df_imputed = pd.concat([df_without_categorical_value,df[["departman","urun_ismi"]]],axis=1)
df_imputed.columns= df.columns 
```

```python
df_imputed
```

```python
%matplotlib inline
df_imputed.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
def make_pie(column):
    values = df_imputed[column].value_counts()
    plt.pie(values.values,labels = values.index,shadow = True,radius=3,autopct='%1.2f%%')
```

```python
make_pie('departman')
```

```python
make_pie('urun_ismi')
```

```python
days = {0.0:'Mon',1.0:'Tues',2.0:'Weds',3.0:'Thurs',4.0:'Fri',5.0:'Sat',6.0:'Sun'}
df_imputed["siparis_gunu"] = df_imputed["siparis_edilen_gun_sayisi"].apply(lambda x : days[x])
```

```python
make_pie('siparis_gunu')
```

```python
df_imputed.groupby('siparis_gunu')["departman"].value_counts().plot(kind="pie",figsize=(30,15))
```

```python
fig, ax = plt.subplots()
ax.set_title('Multiple Samples with Different sizes')
ax.boxplot(df_imputed['siparis_saati'])

plt.show()
```

```python
df_imputed["siparis_numarasi"] =  np.log(df_imputed["siparis_numarasi"])
df_imputed["siparis_gecmisi"] =  np.log(df_imputed["siparis_gecmisi"]+1)
df_imputed["sepete_eklenen_urun_sayisi"] =  np.log(df_imputed["sepete_eklenen_urun_sayisi"]+1)
```

```python
df_imputed.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
target = df_imputed["urun_ismi"]
feature = df_imputed.drop(["urun_id","departman_id","urun_ismi","departman"],axis=1)#it correlates highly with target
```

```python
x_train,x_test,y_train,y_test = train_test_split(feature,target)
```

```python
!pip install flaml
```

```python
from flaml import AutoML
automl_log = AutoML()
```

```python
automl_log.fit(x_train, y_train, task="classification",metric='log_loss',time_budget=60,estimator_list = ['lgbm'],n_splits=5)
```

```python
print('Best log_loss on validation data: {0:.4g}'.format(automl_log.best_loss))
```

```python
import numpy as np
feature_importance = automl_log.feature_importances_
sorted_idx = np.argsort(feature_importance)[-25:]
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 6))
plt.barh(range(len(sorted_idx)), feature_importance[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), np.array(x_train.columns)[sorted_idx])
plt.title('Feature Importance')
```

```python
automl_log.predict(x_test)
```

```python
import pickle
with open('model.pkl','wb') as files:
    pickle.dump(automl_log,files)
```
